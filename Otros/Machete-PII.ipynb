```python
# CLASES
# Crea la clase base "Persona" con los siguientes atributos: nombre - edad 
# Incluir un constructor que inicialice los atributos
class Persona:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad

# Crea una clase "Estudiante" que herede de "Persona" y tenga un atributo adicional "escuela". 
class Estudiante(Persona):
    def __init__(self, nombre,edad,escuela):
        super().__init__(nombre, edad)
        self.escuela = escuela

# Agrega métodos para imprimir la información del estudiante.
    def informacion_estudiante(self):
        print("Nombre:", self.nombre,"  Edad:", self.edad ,"  Escuela:", self.escuela)

# Agrega métodos para imprimir la información del estudiante
estudiante1 =Estudiante("Nahuel",45,"Politecnico Malvinas",)
estudiante1.informacion_estudiante()

# ACTIVIDADES RESUELTAS CLASE 9 
```
